Oh no! A new, deadly virus named dogepox has recently escaped a research lab. Given how dangerous the new strain is, scientists have issued a global epidemic warning, putting the entire planet into panic! They've flown in a team of the world's best data scientists to track the spread of this disease, and you're one of them! 

In this module, you'll be carefully monitoring the spread of dogepox, helping humanity better understand the risk at hand. For this problem, everyone in the world has a unique, non-negative integer id number. Further, all time passes as discrete, integer time steps. In the beginning, only patient zero (with id = 0) has been infected. You will receive an array of interactions (in chronological order), each of which represents two people in the world coming into close proximity on some day.

If two people come together with a distance of less than or equal to 10, and one of them is infected, then the other will become infected immediately. Your task is to output, one number per line, the number of total people infected on each day, starting at day 1, and ending at day 500.

Each interaction is encoded as a dictionary. ids is a pair containing the id's of the two people involved:
```
example_interaction = {
    "ids": (0, 7),
    "day": 3,
    "distance": 25
}
```
Example 1:
```
Input:
interactions = [
    {ids: (0, 1), day: 2, distance: 10},
    {ids: (0, 9), day: 3, distance: 13},
    {ids: (7, 1), day: 5, distance: 8}
]

Output:
1
2
2
2
3
...
```

Explanation:
```
On day 1, only patient zero is infected.
On day 2, person 0 infects person 1.
On day 3, person 9 is safe because the interaction distance was greater than 10.
On day 4, no interactions happen.
On day 5, person 1 infects person 7.
...
```

````python
def infected_timeseries(self, interactions):
    #everyone has unique id non-neg
    #patient zero id=0
    total = 1
    infected = {0: 1}
    yesterday = 1
    for inter in interactions:
        today = inter['day']
        if (today > yesterday):
            for i in range(0,(today - yesterday)):
                print(total)

        if inter['distance'] <= 10:
            first = inter['ids'][0]
            second = inter['ids'][1]
            if infected.get(first) == 1 and infected.get(second) != 1:
                infected[second] = 1
                total += 1
            elif infected.get(first) != 1 and infected.get(second) == 1:
                infected[first] = 1
                total += 1
        yesterday = inter['day']
    print(total) #need this for the final day since the print happens earlier on
````
