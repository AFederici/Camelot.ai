 Scrape the book data in an organized manner from the webpage, do some simple analysis, and decide which ones to buy.

In this problem, you will scrape the data from the website your friend sent: http://books.toscrape.com/catalogue/page-1.html. Follow these steps:
Notice that the books are distributed across several pages, with ~20 per page. You will only be dealing with the first 5 pages. You can figure out how to systematically do this by examining the URLs.
Next, for each page, use BeautifulSoup to scrape the HTML of the page and identify each book.
For each book, obtain the price and and number of stars (the latter is a bit trickier). You'll want to  "Inspect" the HTML of the page to determine the structure (Right click + Inspect). 
Write a function get_books() that prints the price and number of stars for each book, one per line. Only print the books with 4 or 5 stars, from the first 5 pages, in the order they are displayed in the HTML of the website. Check out our simple guide on BeautifulSoup!

````python
from bs4 import BeautifulSoup
import urllib.request

def get_books(self):
    number = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
    for i in range(1,6):
        url = 'http://books.toscrape.com/catalogue/page-' + str(i) +'.html'
        r = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(r, 'html.parser')
        for x in soup.find_all('li', {'class': 'col-xs-6 col-sm-4 col-md-3 col-lg-3'}):
            price =  x.find('p', {'class': 'price_color'}).getText()
            cols = x.find_all('p')
            for col in cols:
                if col.has_key('class') and col['class'][0] == 'star-rating':
                    star = col['class'][1]
            star = star.lower()
            if(number[star]) > 3:
                print(price[1:], number[star])
````

Suppose you wanted to build a model to extend this data to be able to predict the number of stars or future books, so you no longer need to rely on black market data to make money. A commonly used technique is that of decision trees, and random forests.

You may have heard of these before. In this problem, you'll be working with a conceptual step in the models, called entropy. We'll give an overview here, but for more info, see here.

Suppose a variable X (i.e. number of stars, or length of the title) has multiple possible (discrete) outcomes, with probabilities p_1, p_2, ..., p_n. Then the entropy of that variable is E(X) = sum_{i=1}^n - p_i * log_2(p_i). You can think of entropy as inversely related to useful information, so to learn things from the data, you want to decrease entropy. 

Now suppose we want to bring in more features, like the length of the title of the book. We'll set a parameter cutoff, and encode books that have a title with strictly more characters than cutoff with 1, and 0 otherwise. We now have two variables. To get the new entropy, we iterate across the outcomes for number of stars, and take the weighted sum of the entropies of the title length: sum_{i=1}^5 p_i E(X_i). Here, X_i indicates the data points where the number of stars equals i. See the example for more details.

You will write a function get_entropy(df, cutoff), that takes in a Pandas DataFrame containing the data scraped from the previous problem, with the addition of the book title. Your job is to print the entropy of the 2 fields: E(number of stars, length of title). 
```
Example 1:
Input:
df = 
                                    name  price  stars
0                   A Light in the Attic  51.77      3
1                     Tipping the Velvet  53.74      1
2                             Soumission  50.10      1
3                          Sharp Objects  47.82      4
4  Sapiens: A Brief History of Humankind  54.23      5
...

cutoff = 12

Output:
<entropy>

Explanation:
Suppose 'var1' is either A or B, and 'var2' is either 0 or 1. Suppose that var1 is equally likely to be A or B. When var1 is A, var2 is 0 only 1/3 of the time. When var2 is B, var2 is 0 3/4 of the time. So the entropy E(var1, var2) is: 

(1/2) * (-(1/3)log2(1/3) - (2/3)log2(2/3)) + (1/2) * (-(3/4)log2(3/4) - (1/4)log2(1/4))
```

````python
#This solution is rather poor code; however, I wrote it as a part of a timed competition and went about it in a simple way.
#My more elegant solution failed a single test case, but I am confident it would be easy to convert all my duplicated code into 1 simpler function

import pandas as pd
import numpy as np
from collections import Counter
######################################################################
#
# You can compute a base 2 logarithm with:
# np.log2(x)
#
######################################################################
def entropy_f(prob):
    return -1.0 * prob * np.log2(prob)
class Solution:
    def get_entropy(self, df, cutoff):
        df['e'] = 0
        for ind, row in df.iterrows():
            if len(row['name']) > cutoff:
                #print(len(row['name']))
                df['e'][ind] = 1
        list_stars = df['stars'].tolist()
        s1_0 = 0
        s2_0 = 0
        s3_0 = 0
        s4_0 = 0
        s5_0 = 0
        list_title = df['e'].tolist()
        total_len = len(list_stars)
        for i in range(total_len):
            if list_stars[i] == 1:
                if list_title[i] == 0:
                    s1_0 += 1
            elif list_stars[i] == 2:
                if list_title[i] == 0:
                    s2_0 += 1
            elif list_stars[i] == 3:
                if list_title[i] == 0:
                    s3_0 += 1
            elif list_stars[i] == 4:
                if list_title[i] == 0:
                    s4_0 += 1
            elif list_stars[i] == 5:
                if list_title[i] == 0:
                    s5_0 += 1
        val_stars = Counter(list_stars)
        for key in val_stars:
            if key == 1:
                prob_s1 = 1.0 * val_stars[key] / total_len
                prob_t1 = 1.0 * s1_0 / val_stars[key]
            if key == 2:
                prob_s2 = 1.0 * val_stars[key] / total_len
                prob_t2 = 1.0 * s2_0 / val_stars[key]
            if key == 3:
                prob_s3 = 1.0 * val_stars[key] / total_len
                prob_t3 = 1.0 * s3_0 / val_stars[key]
            if key == 4:
                prob_s4 = 1.0 * val_stars[key] / total_len
                prob_t4 = 1.0 * s4_0 / val_stars[key]
            if key == 5:
                prob_s5 = 1.0 * val_stars[key] / total_len
                prob_t5 = 1.0 * s5_0 / val_stars[key]
        entropy = prob_s1 * (entropy_f(prob_t1) + entropy_f(1- prob_t1))
        entropy += prob_s2 * (entropy_f(prob_t2) + entropy_f(1- prob_t2))
        entropy += prob_s3 * (entropy_f(prob_t3) + entropy_f(1- prob_t3))
        entropy += prob_s4 * (entropy_f(prob_t4) + entropy_f(1- prob_t4))
        entropy += prob_s5 * (entropy_f(prob_t5) + entropy_f(1- prob_t5))
        print('%.5f'%(entropy))

````

Now it's time to go profit when the bookstore opens. Luckily, the bookstore is tech savvy, and posts all the books and prices online. You still have your database of book listings to figure out which ones to buy. On average, you can expect to make 2 dollars for a 5 star book, 1 dollar for a 4 star book, none for 3 stars, lose 1 dollar for 2 stars, and lose 2 dollars for 1 star. There will be some noise in the amount of profit. 

Unfortunately there is a bug in the bookstore's system, and you're only able to see the first word of the title! Fortunately, this is usually enough for you to identify the book.

Write a function buy_books(df, first_words), that takes in the same Pandas DataFrame as before, along with a list of book titles' first-words. Use the data to determine which books to buy. Print your decision for each book, one per line, with 1 indicating buy, and 0 otherwise. Your score will be your total profit (or loss). 
```
Input:
df = <same_as_before>,
books = ['Walls', 'The', ...]

Output:
1
0
0
1
...
```

````python
import pandas as pd
import numpy as np
def split(s):
    return s.split()[0]
def money(s):
    profits = {1:-2, 2:-1, 3:0, 4:1, 5:2}
    return profits[s]
class Solution:
    def buy_books(self, df, first_words):
        df['name'] = df['name'].apply(split)
        df['profits'] = df['stars'].apply(money)
        df2 = df.groupby('name')['profits'].sum()
        for word in first_words:
            if df2[word] > 0:
                print(1)
                
            else:
                print(0)
````
